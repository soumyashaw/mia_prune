#### Plot 8.1.1

```python
## CIFAR 10, ResNet18

data = [("L1 Unstructured", 0.5, 69.63), ("L1 Unstructured", 0.6, 70.467), ("L1 Unstructured", 0.7, 70.511), ("L1 Unstructured", 0.8, 69.089), ("L1 Unstructured", 0.9, 68.319),
        ("L1 Structured", 0.5, 67.926), ("L1 Structured", 0.6, 68.733), ("L1 Structured", 0.7, 67.822), ("L1 Structured", 0.8, 65.207), ("L1 Structured", 0.9, 58.037),
        ("L2 Structured", 0.5, 70.133), ("L2 Structured", 0.6, 68.830), ("L2 Structured", 0.7, 67.459), ("L2 Structured", 0.8, 64.585), ("L2 Structured", 0.9, 56.156)]

df = pd.DataFrame(data=data, columns=["pruner_name", "prune_sparsity", "prediction_accuracy"])

#print(df)

sns.set_theme(style="darkgrid", palette="bright", context="notebook")
sns.set_style(rc={"axes.grid": True, "grid.linestyle": "--"})
ax = sns.lineplot(
    data=df,
    x="prune_sparsity", y="prediction_accuracy", hue="pruner_name", style="pruner_name",
    markers=["*"], dashes=False, markersize = 10
)
ax.axhline(67.696, color='black', linestyle='--', label="Original Model")
ax.set(xlabel='Model Sparsity', ylabel='Prediction Accuracy (%)')#, ylim=(10, 80))

plt.legend(title='Pruner Name')
plt.show()
```
